## Importing Libraries

```python
import operator
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
import matplotlib.pyplot as plt

from helper import get_text_length, get_freq_words, get_uniq_words, get_most_freq
```

----------------------------------------------
# TRAIN DATASET
----------------------------------------------

## Loading Training Set

```python
train_df = pd.read_csv("data/train_data.csv")

train_df
```

## Checking for Missing Values

```python
train_df.isna().sum()
```

## Information about the Dataset

```python
train_df.info()
```

## Calculating the Length of Text

```python
train_df["text_length"] = train_df["input"].apply(get_text_length)

train_df
```

## Statistics about Text Lengths

```python
text_lengths = train_df["text_length"]
print("min length :", min(text_lengths))
print("max length :", max(text_lengths))
print("average length :", np.average(text_lengths))
print("std length :", np.std(text_lengths))
text_lengths.describe()
```

**Observation:** from the above statistics we observe min length is 2, max length is 319 and average length 11.725

## PDF of Text Lengths

```python
plt.figure(figsize=(12,8))
sns.distplot(text_lengths, bins=90)
plt.xlabel('text_length', fontsize=12)
plt.title("PDF of text_lengths")
```

## Frequency of Words

```python
t_word_count = get_freq_words(train_df["input"])
t_word_count
```

## Finding Number of Unique Words

```python
t_uniq_words = get_uniq_words(t_word_count)
print("Number of unique words : ", len(t_uniq_words))
```

## Displaying top 100 most Frequent Words

```python
t_most_freq = get_most_freq(t_uniq_words)
print(t_most_freq)
```
